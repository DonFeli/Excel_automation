```python
# Chargement des librairies

from openpyxl import Workbook
from openpyxl import load_workbook
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from openpyxl.utils.dataframe import dataframe_to_rows
from openpyxl.utils import FORMULAE
from openpyxl.formula.translate import Translator
from openpyxl.chart import BarChart, Reference
```

```python
# Chargement du fichier
workbook = load_workbook(filename="avions.xlsx")
```

```python
# Afficher les noms des feuilles
workbook.sheetnames
```

```python
##affectation
sheet = workbook['Feuil1']
sheet
sheet.title
```

```python
##creation d'un dataframe
df = pd.DataFrame(sheet.values)

##formatage des données
df.columns = df.iloc[0]

df=df.drop([0, 1])
```

```python
##vérification des données
df.head()

list(df.columns)
```

```python
##préparation des données pour classification
X= df[['Seat comfort', 'Departure/Arrival time convenient', 'Food and drink', 'Gate location', 'Inflight wifi service','Inflight entertainment','Online support','Ease of Online booking','On-board service','Leg room service','Baggage handling','Checkin service','Cleanliness','Online boarding']]
```

```python
##normalisation
X_norm=(X-X.mean())/X.std()

X_col_name=list(X_norm.columns)
```

```python
###definition du nombre de classe par la méthode elbow
Sum_of_squared_distances = []
K = range(1,10)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(X_norm)
    Sum_of_squared_distances.append(km.inertia_)

##nombre de cluster
plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
##transformation en matrice
X_norm=X_norm.to_numpy()

#Cluster K-means
model=KMeans(n_clusters=5).fit(X_norm)

#adapter le modèle de données 
###attention fit(X) avec labels_ est exactement la même chose que model.predict(X)
labels=model.predict(X_norm)
print(model.labels_)

##classes
print('centres des classes')
print(model.cluster_centers_)
```

```python
##profil type de chaque classe
profil_type= pd.DataFrame(model.cluster_centers_, columns=X_col_name)
```

```python
##plot profil type
ax = profil_type.plot.bar(figsize=(150,150))
plt.legend(prop={'size':70})
```

```python
##ajout de la classe pour chaque individu
df['classification'] = pd.Series(model.labels_, df.index)
```

```python
##creation d'une feuille par classe
arraysDict = {}
list_name = []
for i in range(0,5):
    name = "classe"+str(i)
    list_name.append(name)
    arraysDict['classe{0}'.format(i)] = workbook.create_sheet(name)
```

```python
##dans chaque feuille on ajoute les individus de la classe
for i in range(0,5):
    df_fil=df[df['classification']==i]
    name="classe"+str(i)
    sheet2 = workbook[name]
    for r in dataframe_to_rows(df_fil, index=False, header=False):
        sheet2.append(r)
```

```python
##création d'une feuille de reporting pour chaque classe
arraysDict1 = {}
list_name_stats = []
for i in range(0,5):
    name = "classe"+str(i)+"_stats"
    list_name_stats.append(name)
    arraysDict1['classe{0}'.format(i)] = workbook.create_sheet(name)
```

```python
list_name_stats
```

```python
##dictionnaire des variables
dict_var={'Seat comfort', 'Departure/Arrival time convenient', 'Food and drink', 'Gate location', 'Inflight wifi service','Inflight entertainment','Online support','Ease of Online booking','On-board service','Leg room service','Baggage handling','Checkin service','Cleanliness','Online boarding'}
```

```python
def scoring(df) :   
    score=df['Seat comfort']+df['Departure/Arrival time convenient']+df['Food and drink']+df['Gate location']+df['Inflight wifi service']+df['Inflight entertainment']+df['Online support']+df['Ease of Online booking']+df['On-board service']+df['Leg room service']+df['Baggage handling']+ df['Checkin service']+df['Cleanliness']+df['Online boarding']
    return score
```

```python


j=0
for i in list_name_stats:
        print(i)
        ##variable dans laquelle on met la feuille dans laquelle on va implémenter
        sheet3= workbook[i]
        ##ajout des cellules textes
        sheet3["A1"].value = "nombre d'individus"
        sheet3["B1"].value = "nombre d'individus satisfait"
        sheet3["C1"].value = "nombre d'individus non satisfait"


        sheet3["B5"].value = "moyenne"
        sheet3["C5"].value = "mediane"
        sheet3["D5"].value = "ecart_type"
        sheet3["E5"].value = "nbr_0"
        sheet3["F5"].value = "nbr_5"

        sheet3["A6"].value='Seat comfort'
        sheet3["A7"].value='Departure/Arrival time convenient'
        sheet3["A8"].value='Food and drink'
        sheet3["A9"].value='Gate location'
        sheet3["A10"].value='Inflight wifi service'
        sheet3["A11"].value='Inflight entertainment'
        sheet3["A12"].value='Online support'
        sheet3["A13"].value='Ease of Online booking'
        sheet3["A14"].value='On-board service'
        sheet3["A15"].value='Leg room service'
        sheet3["A16"].value='Baggage handling'
        sheet3["A17"].value='Checkin service'
        sheet3["A18"].value='Cleanliness'
        sheet3["A19"].value='Online boarding'


        sheet3["A23"].value='score_total'
        sheet3["A24"].value='>30'
        sheet3["A25"].value='30 à 40'
        sheet3["A26"].value='40 à 50'
        sheet3["A27"].value='<50'
        sheet3["B23"].value='satisfait'
        sheet3["C23"].value='non satisfait'
        sheet3["D23"].value='raiting'
        
        ##dataframe de la classe dans laquelle nous faisons notre implémentation
        df_temp=df[df['classification']==j]
        ##feuille excel avec les individus de la classe
        feuil= list_name[j]
        
        ##calculs excel
        formule = """=NB.SI(""" + feuil + """!B2:"""+feuil +"""!B1999,"=0")"""
        sheet3["B2"].value=formule
        formule = """=NB.SI(""" + feuil + """!B2:"""+feuil +"""!B1999,"=1")"""
        sheet3["C2"].value=formule
        sheet3["A2"].value= """= B2+ C2  """
        
        ###implementation du tableau de stats descriptive
        for cell in (['B', 'C', 'D', 'E', 'F']):
            k=6
            ###boucle sur chaque variable
            for var in dict_var:
                if cell=='B':
                    ##moyenne
                    test='B'+str(k)
                    sheet3[test]=df_temp[var].mean()
                    k=k+1
                if cell=='C':
                    ##mediane
                    test='C'+str(k)
                    sheet3[test]=df_temp[var].median()
                    k=k+1
                if cell=='D':
                    ##ecart_type
                    test='D'+str(k)
                    sheet3[test]=df_temp[var].std()
                    k=k+1
                if cell=='E':
                    ##nombre de valeur min
                    df_temp2= df_temp[df_temp[var]==0]
                    test='E'+str(k)
                    sheet3[test]=len(df_temp2)
                    k=k+1
                if cell=='F':
                    ##nombre de valeur max
                    df_temp2= df_temp[df_temp[j]==5]
                    test='F'+str(k)
                    sheet3[test]=len(df_temp2)
                    k=k+1
                    
                    
        df_temp['score']=scoring(df_temp)
        ###inférieur à 30
        sheet3["B24"].value=df_temp[(df_temp.satisfaction=='satisfied') &  (df_temp.score<=30)].satisfaction.count()
        sheet3["C24"].value=df_temp[(df_temp.satisfaction=='dissatisfied') & (df_temp.score<=30)].satisfaction.count()
        if df_temp[(df_temp.satisfaction=='satisfied') &  (df_temp.score<=30)].satisfaction.count() > 0:
            sheet3["D24"].value= df_temp[(df_temp.satisfaction=='satisfied') &  (df_temp.score<=30)].satisfaction.count() / df_temp[(df_temp.score<=30)].satisfaction.count()*100
        else:
            sheet3["D24"].value=0

        ###30 à 40
        sheet3["B25"].value=df_temp[(df_temp.satisfaction=='satisfied') &  (df_temp.score>30) & (df_temp.score<=40)].satisfaction.count()
        sheet3["C25"].value=df_temp[(df_temp.satisfaction=='dissatisfied') & (df_temp.score>30) & (df_temp.score<=40)].satisfaction.count()
        if df_temp[(df_temp.satisfaction=='satisfied') &  (df_temp.score>30) & (df_temp.score<=40)].satisfaction.count() > 0:
            sheet3["D25"].value= df_temp[(df_temp.satisfaction=='satisfied') &  (df_temp.score>30) &(df_temp.score<=40)].satisfaction.count() / df_temp[(df_temp.score>30)& (df_temp.score<=40)].satisfaction.count()*100
        else:
            sheet3["D25"].value=0
    
        ###supérieur à 40
        sheet3["B26"].value=df_temp[(df_temp.satisfaction=='satisfied') &  (df_temp.score>40) & (df_temp.score<=50)].satisfaction.count()
        sheet3["C26"].value=df_temp[(df_temp.satisfaction=='dissatisfied') & (df_temp.score>40) & (df_temp.score<=50)].satisfaction.count()
        if df_temp[(df_temp.satisfaction=='satisfied') &  (df_temp.score>40) & (df_temp.score<=50)].satisfaction.count() > 0:
            sheet3["D26"].value= df_temp[(df_temp.satisfaction=='satisfied') &  (df_temp.score>40) & (df_temp.score<=50)].satisfaction.count() / df_temp[(df_temp.score>40) & (df_temp.score<=50)].satisfaction.count()*100
        else:
            sheet3["D26"].value=0
    
        ###supérieur à 50
        sheet3["B27"].value=df_temp[(df_temp.satisfaction=='satisfied') &  (df_temp.score>50)].satisfaction.count()
        sheet3["C27"].value=df_temp[(df_temp.satisfaction=='dissatisfied') & (df_temp.score>50)].satisfaction.count()
        if df_temp[(df_temp.satisfaction=='satisfied') &  (df_temp.score>50)].satisfaction.count() > 0:
            sheet3["D27"].value= df_temp[(df_temp.satisfaction=='satisfied') &  (df_temp.score>50)].satisfaction.count() / df_temp[(df_temp.score>50)].satisfaction.count()*100
        else:
            sheet3["D27"].value=0
            
        ##graph1
        graph = BarChart()
        data = Reference(sheet3, min_row=5, max_row=19, min_col=2, max_col=2)
        cats = Reference(sheet3, min_col=1, min_row=6, max_row=20)


        graph.add_data(data, titles_from_data=True)
        graph.set_categories(cats)
        sheet3.add_chart(graph, "I2")

        ##graph2
        graph2 = BarChart()
        data2 = Reference(sheet3, min_row=23, max_row=27, min_col=4, max_col=4)
        cats2 = Reference(sheet3, min_col=1, min_row=24, max_row=27)


        graph2.add_data(data2, titles_from_data=True)
        graph2.set_categories(cats2)
        sheet3.add_chart(graph2, "I23")   
        
        j=j+1
```
